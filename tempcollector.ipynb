```python
import os
import json
```

```python
root_path = "./works"
total = dict()
for file in os.listdir(root_path):
    if file.endswith(".txt"):
        with open(os.path.join(root_path, file), "r", encoding="utf-8") as f:
            try:
                total[file[:-4]] = json.load(f)
            except json.JSONDecodeError:
                print(file)
```
